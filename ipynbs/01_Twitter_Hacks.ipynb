```python
import json
query_term = "politics"
search_results = [tweeter.search(q=query_term, rpp=100, page=page) for page in range(1,2)]
print json.dumps(search_results, sort_keys=True, indent=1)
```

<h2>Frequency Analysis & Lexical Diversity</h2>
Lexical diversity: the number of unique tokens in a piece of text divided by the total numbe of tokens
Example 1-7 (page 7)

```python
tweets = [r \
             for result in search_results \
                     for r in result['results']]

words = [w \
           for t in tweets \
              for w in t['text'].split()]
total_words = len(words)
unique_words = len(set(words))
diversity = unique_words/float(total_words)
avg_words_tweet = sum([len(t['text'].split()) for t in tweets])/float(len(tweets))

print 'The lexical diversity is {0}'.format(diversity)
print 'The average number of words per tweet is {0}'.format(avg_words_tweet)
```

<h2>Frequency Distribution Using NLTK</h2>

```python
import nltk
import math
freq_dist = nltk.FreqDist(words)
n = min(50, len(freq_dist.keys())/2)
print 'The {0} most popular words are\n{1}\n'.format(n, freq_dist.keys()[:n])
print 'The {0} least popular words are\n{1}'.format(n, freq_dist.keys()[-n:])
```

<h2>Retweet Graph Using Networkx</h2>

```python
import networkx as nx
import re
g = nx.DiGraph()

#get retweet sources
def get_rt_sources(tweet):
    rt_patterns = re.compile(r"(RT|via)((?:\b\W*@\w+)+)",re.IGNORECASE)
    return [source.strip()
              for tuple in rt_patterns.findall(tweet)
                  for source in tuple
                     if source not in ("RT", "via") ]
#tweets from Frequency Analysis & Lexical Diversity example
for tweet in tweets:
    rt_sources = get_rt_sources(tweet["text"])
    if not rt_sources: continue
    for rt_source in rt_sources:
        #from_user is the retweeter
        #rt_source is the original tweeter
        #so this adds an edge that points from the original tweeter to the user that retweeted
        g.add_edge(rt_source, tweet["from_user"], {"tweet_id" : tweet["id"]})

print "The graph has {0} nodes".format(g.number_of_nodes())
print "The graph has {0} edges".format(g.number_of_edges())
print "There are {0} subgraphs".format(len(nx.connected_components(g.to_undirected())))
print "The highest degree node has degree {0}".format(max(nx.degree(g).values()))
```

<h2>Visualizing Tweet Graphs</h2>

```python
OUT = "example_1_12.out"
try:
    nx.drawing.write_dot(g, OUT)
except ImportError, e:
    print e
#system command to convert this dot file to a png
import subprocess
PNG = "example_1_12"
subprocess.call(["circo", "-Tpng", "-O{0}".format(PNG), OUT])
```
